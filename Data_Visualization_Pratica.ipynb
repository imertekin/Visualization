```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Product.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
data=df.drop(['id','Product_img_url','Product_url','Product_outsize','Product_insize'],axis=1)
data['Add_date']=data['Add_date'].astype('datetime64[ns]')
```

```python
data.head()
```

```python
data.Product_Free_Shipping=[1 if i==False else 0 for i in data.Product_Free_Shipping]

```

```python
uniq_cata=pd.DataFrame(data.Product_Dsc_Price.groupby(by=data.Product_MinCatagory).max())
uniq_cata=uniq_cata.reset_index()
uniq_cata.head()
```

```python
plt.figure(figsize=(20,10))
sns.set_style("whitegrid")
sns.barplot(x=uniq_cata.Product_MinCatagory ,y=uniq_cata.Product_Dsc_Price)
plt.xticks(rotation=90)
plt.show()
```

```python
t_shirt=data[data.Product_MinCatagory=='T-Shirt']
t_shirt.head()
```

```python
tshirt_brand=pd.DataFrame(t_shirt.Product_Dsc_Price.groupby(by=t_shirt.Product_Brand).max())
tshirt_brand=tshirt_brand.reset_index()
tshirt_brand.head()
```

```python
plt.figure(figsize=(20,5))
sns.barplot(x=tshirt_brand.Product_Brand ,y=tshirt_brand.Product_Dsc_Price)
plt.xticks(rotation=90)
plt.show()
```

```python
brand_list=list(t_shirt.Product_Brand.value_counts().head().index)


m_tshirt=t_shirt[t_shirt.Product_Brand.isin(brand_list)]
    
m_tshirt['Product_Brand'].value_counts()
```

```python

plt.figure(figsize=(10,10))
plt.plot(m_tshirt.Product_Dsc_Price.groupby(by=m_tshirt.Product_Brand).mean())
sns.scatterplot(x=m_tshirt.Product_Brand,y=m_tshirt.Product_Dsc_Price,hue=m_tshirt.Product_Ratings)
```

```python
plt.figure(figsize=(15,10))

sns.lineplot(x=m_tshirt['Add_date'].dt.to_period('D').astype('string').astype('datetime64'),y=m_tshirt.Product_Dsc_Price)

plt.show()
```

```python
tshirt_mean=t_shirt['Product_Dsc_Price'].groupby(t_shirt['Add_date'].dt.to_period('D')).mean()
tshirt_mean=tshirt_mean.reset_index()
tshirt_mean['Add_date']=tshirt_mean.Add_date.astype('string').astype('datetime64')
tshirt_mean.head()
```

```python
plt.figure(figsize=(15,10))

sns.lineplot(x=tshirt_mean.Add_date,y=tshirt_mean.Product_Dsc_Price)
```
